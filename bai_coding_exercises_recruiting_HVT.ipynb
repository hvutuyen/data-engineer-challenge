### **Aufgabe 3**
a) Obwohl die Spalte `price` numerische Werte besitzt, ist sie vom Datentyp *object*. Dies liegt einerseits am Sonderzeichen `€` sowie dem Dezimaltrennzeichen `,` andererseits. Lösche daher das `€`-Zeichen und ersetze das Dezimaltrennzeichen `,` durch `.`. Caste die Spalte im Anschluss in den Datentyp *float*.  
b) In der Spalte `price` fehlen zwei Werte, sog. *NaN*s. Ersetze die fehlenden Werte mit dem Preis des jeweiligen Produkts. (Es existieren keine Preisunterschiede zwischen den Filialen) 

```python
# Loesungen Aufgabe 3

# a) 
def convert_price_column(df: pd.DataFrame, column_name: str = "price") -> pd.DataFrame:
    """
    removes the "€" character, replaces "," with "."
    and converts the column into a float datatype

    Args:
        - Dataframe
        - Column_name which will be processed (default set to "price")

    Returns:
        Dataframe with converted price column
    """
    
    # remove "€"
    df[column_name] = df[column_name].str.replace("€", "", regex=False) # regex = False, since the strings are simple -> no pattern
    
    # replace "," with "."
    df[column_name] = df[column_name].str.replace(",", ".", regex=False)
    
    # convert column type to float
    df[column_name] = df[column_name].astype(float)

    return df

# call function to convert price column
converted_price_df = convert_price_column(delivery_data_with_no_duplicates)

# check the datatype of "price" column
print(converted_price_df.dtypes["price"])
#print(converted_price_df.tail())

# b) fillna()
def fill_nan_for_price_col(df: pd.DataFrame, value_column: str ="price", group_column: str = "id_product") -> pd.DataFrame:
    """
    fills NaN for the "price" column with the corresponding values of the same "id_product"

    Args:
        - Dataframe
        - value_column as string (default set to "price")
        - group_column as string (default set to "id_product")

    Returns:
        - DataFrame with the filled NaNs
    """

    # fill NaNs in value_column with the median value of the group ("id_product")
    df[value_column] = df.groupby(group_column)[value_column].transform(lambda x: x.fillna(x.median()))
    
    return df

# call function
#converted_price_df = converted_price_df.copy()
filled_price_df = fill_nan_for_price_col(converted_price_df)

# check if there are still any NaN in the "price" column
has_nan = filled_price_df['price'].isna().any()
if has_nan:
    print("Still has NaN!")
else:
    print("No NaNs found")

filled_price_df.tail()
```

### **Aufgabe 4**
a) Zu den bereits eingelesenen Liefermengen existieren auch Verkaufsmengen, die sich in der Datei `sales.xlsx` befinden. Lese die Datei ein.  
b) Merge die Liefer- und die soeben eingelesenen Verkaufszahlen auf den Spalten `id_store`, `id_product` und `delivery_date`/`sale_date`.  
c) In welcher der beiden Filialen (`id_store`) war der Umsatz am 24.05.2023 höher?

```python
# Loesungen Aufgabe 4

# a) read salex.xlsx
sales_df = pd.read_excel("sales.xlsx", sheet_name=0) # 0 == first sheet
#sales_df.head()
#print(sales_df.dtypes)

# b) merge
# assuming that we want to know how many items were sold on the same day as the delivery day -> merge "delivery_date" col with "sale_date" col
merged_df = pd.merge(filled_price_df, sales_df, left_on=["id_store", "id_product", "delivery_date"], right_on=["id_store", "id_product", "sale_date"])
#print(merged_df.head())

# c)
def determine_store_with_highest_revenue(df: pd.DataFrame, date: str = "2023-05-24") -> str:
    """
    determines the store with the highest revenue for a specific date
    
    Args:
        - Dataframe which is already merged with the sales data
        - date as string in format YYYY-MM-DD (default set to 2023-05-24)

    Returns:
        - Sentence saying which store had the highest revenue
    """

    # filter rows where "sale_date" is the defined date
    filtered_date_df = df[df['sale_date'] == date]

    filtered_date_df = filtered_date_df.copy()

    # create new column "revenue" which equals the product of "sales_qty" and "price"
    filtered_date_df["revenue"] = filtered_date_df["sales_qty"] * filtered_date_df["price"]

    # groub by "id_store" and aggregate (sum) of "revenue"
    revenue_by_store = filtered_date_df.groupby("id_store")["revenue"].sum().reset_index()
    
    # sort by "revenue" descending 
    sorted_revenue_by_store = revenue_by_store.sort_values("revenue", ascending=False) # first row has the highest revenue

    # access first row and store information in variables
    highest_revenue_row = sorted_revenue_by_store.iloc[0]
    store_id = highest_revenue_row["id_store"]
    rev = highest_revenue_row["revenue"]

    # return statement
    return f"Store {int(store_id)} hat mit {rev}€ am {date} den höchsten Umsatz generiert."

# assign und dann max auf die spalte -> selektiert direkt den höchsten wert

# call function to determine the store with the highest revenue on 24.05.2023
print(determine_store_with_highest_revenue(merged_df))
```

### **Aufgabe 5**
a) Bei der Benamung des *"KRaichgauer Spargelbrot 0,5kg"* scheint ein Fehler unterlaufen zu sein. Ersetze das große *R* an zweiter Stelle durch ein kleines ("K*R*aichgauer Spargelbrot 0,5kg" --> "K*r*aichgauer Spargelbrot 0,5kg")  
b) Neben dem Namen gibt die Spalte `name_product` auch Auskunft über das Gewicht des jeweiligen Produkts. Extrahiere das Gewicht und schreibe es in eine neue Spalte `product_weight`, die Informationen über das Produktgewicht in Gramm liefern soll.  
c) Speicherplatz ist teuer. Daher sollen absofort kürzere Artikelnamen eingeführt werden. Lösche alle Zeichen, die nach dem ersten Wort in Spalte `name_product` folgen. (Bsp: *"Kraichgauer Spargelbrot 0,5kg"* --> *"Kraichgauer"*).

```python
# Loesungen Aufgabe 5

# a) 
def correct_string(df: pd.DataFrame, column_name: str="name_product") -> pd.DataFrame:
    """
    corrects capitalization of strings
    
    Args:
        - Dataframe which contains the column that will be corrected
        - column_name as string (default set to name_product)
        
    Returns:
        Dataframe with the corrected column
    """
    
    # convert string to lower case first, then capitalize every new word
    df[column_name] = df[column_name].apply(lambda x: ' '.join(word.capitalize() for word in x.lower().split()))
    
    return df

# call function
corrected_col_df = correct_string(merged_df)
#corrected_col_df.tail()

# b)
def extract_digits_from_string(text: str) -> float:
    """
    extracts all digits from a text and returns them as a string
    
    Args:
        - text as string
    
    Returns:
        - float: digits as float from the extracted text
    """
    digits_str = ''.join([c for c in str(text) if c.isdigit()])
    return float(digits_str) if digits_str else None  # None if no digit was found

corrected_col_df['product_weight'] = corrected_col_df['name_product'].apply(extract_digits_from_string)
#print(corrected_col_df)

# this function only extracts digits but it cannot distinguish between g and kg, so "product_weight" will contain 0.5(kg)
# since "product_weight" should be in grams 0.5(kg) needs to be converted to gram
# possible solution -> RegEx 


# c)
def shorten_product_name(df: pd.DataFrame, column_name: str="name_product") -> pd.DataFrame:
    """
    overwrites column of dataframe, so only the first word of each row will be kept
    
    Args:
        - Dataframe
        - column_name which will be processed (default set to "name_product")
    
    Returns:
        - Dataframe with the processed column
    """
    
    # overwrite column with the first word from the string
    df[column_name] = df[column_name].str.split().str[0] # split to the first whitespace and set entry of the column to that elemental
    
    return df

# call function
shortened_product_name_df = shorten_product_name(corrected_col_df)

# print df with the shortened name_product
shortened_product_name_df.head()
#print(shortened_product_name_df)
```
